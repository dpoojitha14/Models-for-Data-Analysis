Referred from http://rischanlab.github.io/Kmeans.html

```r
#Iris Data
head(iris)
colnames(iris)
```

```r
#Assigning features to the variable x and class (Species) to variable y
x = iris[,-5]
y = iris$Species
```

```r
#No of clusters for the Iris data set = 3
kc = kmeans(x,3)
```

```r
#Summary
kc
```

```r
table(y,kc$cluster)
```

```r
plot(x[c("Sepal.Length", "Sepal.Width")], col=kc$cluster)
points(kc$centers[,c("Sepal.Length", "Sepal.Width")], col=1:3, pch=23, cex=3)
```
